```python
# 단순선형회귀분석

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv('Elderly_Accident_Dataset.csv')
```

```python
# Car 와 ED_NA_2019 간의 회귀분석 및 산점도 시각화
X = df.Car.values
Y = df.ED_NA_2019.values

# 독립 변수 X에 상수항 추가
X = sm.add_constant(X)

# 선형 회귀 모델 적합
model = sm.OLS(Y, X)
result = model.fit()

# 회귀 분석 결과 출력
print(result.summary())

# 산점도와 회귀선 그리기
plt.scatter(X[:, 1], Y, alpha=0.5, c=['red' if x == 498 else 'blue' for x in df.ED_NA_2019])
plt.plot(X[:, 1], result.predict(), color='r')

# 그래프 축 라벨링 및 타이틀 추가
plt.xlabel('Car')
plt.ylabel('ED_NA_2019')
plt.title('Linear Regression')

# 그래프 출력
plt.show()
```
