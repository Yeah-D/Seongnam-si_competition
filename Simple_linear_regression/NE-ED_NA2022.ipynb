```python
# 단순선형회귀분석

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv('G_Elderly_Accident_Dataset.csv')
```

```python
# NE_area 와 NA_area 간의 회귀분석 및 산점도 시각화
X = df.NE_area.values
Y = df.NA_area.values

# 독립 변수 X에 상수항 추가
X = sm.add_constant(X)

# 선형 회귀 모델 적합
model = sm.OLS(Y, X)
result = model.fit()

# 회귀 분석 결과 출력
# print(result.summary())

# 산점도와 회귀선 그리기
plt.scatter(X[:, 1], Y, alpha=0.5, c=['red' if x == 0.00094632049 else 'blue' for x in df.NE_area])
plt.plot(X[:, 1], result.predict(), color='r')

# 그래프 축 라벨링
plt.title('NE/area_2022 ~ NA/area_2022')
plt.xlabel('NE_area_2022')
plt.ylabel('NA_area_2022')

# 그래프 출력
plt.show()

corr, pval = stats.spearmanr(df['NE_area'], df['NA_area'])
print(f"Correlation: {corr:.3f}\nP-value: {pval:.20f}")
```
