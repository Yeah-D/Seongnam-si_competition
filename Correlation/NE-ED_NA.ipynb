```python
# 상관관계분석

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Elderly_Accident_Dataset.csv')

# 자료구조
print(df)
```

```python
# NE 와 ED_NA_2019 간의 상관관계분석 및 산점도 시각화
X = df.NE.values
Y = df.ED_NA_2019.values

plt.scatter(X, Y, alpha=0.5, c=['red' if z == 134023 else 'blue' for z in X])

plt.title('ED_NA_2019 ~ NE')
plt.xlabel('NE')
plt.ylabel('ED_NA_2019')
plt.show()

import scipy.stats as stats

corr, pval = stats.spearmanr(df['NE'], df['ED_NA_2019'])
print(f"Correlation: {corr:.3f}\nP-value: {pval:.20f}")
```
