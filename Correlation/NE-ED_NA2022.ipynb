```python
# 상관관계분석
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('G_Elderly_Accident_Dataset.csv')

# 자료구조
print(df)

# NE_area 와 NA_area 간의 상관관계분석 및 산점도 시각화
X = df.NE_area.values
Y = df.NA_area.values

plt.scatter(X, Y, alpha=0.5, c=['red' if z == 0.00094632049 else 'blue' for z in X])

plt.title('NE/area_2022 ~ NA/area_2022')
plt.xlabel('NE/area_2022')
plt.ylabel('NA/area_2022')
plt.show()

import scipy.stats as stats

corr, pval = stats.spearmanr(df['NE_area'], df['NA_area'])
print(f"Correlation: {corr:.3f}\nP-value: {pval:.20f}")
```
